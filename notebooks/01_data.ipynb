```python
import pandas as pd

df = pd.read_parquet("../data/raw/Массив Данных в формате Parquet/Показатели по регионам.parquet")
print(f"Размер: {df.shape}")
print(f"Колонки: {df.columns.tolist()}")
df.head()
```

```python
import os
files = os.listdir("../data/raw")
for f in files:
    print(repr(f))
```

```python
target_regions = ['Московская область', 'Краснодарский край', 'Республика Саха (Якутия)']
for r in target_regions:
    n = (df['object_name'] == r).sum()
    print(f"{r}: {n} строк")
```

```python
otgr = df[df['indicator_name'].str.contains('тгрузк', case=False, na=False)]
print("=== Отгрузка ===")
print(otgr[['indicator_code', 'indicator_name']].drop_duplicates().head(10).to_string())


shx = df[df['indicator_name'].str.contains('сельск', case=False, na=False)]
print("\n=== С/х ===")
print(shx[['indicator_code', 'indicator_name']].drop_duplicates().head(10).to_string())
```

```python
indicator_code = "Y123"  
region = "Московская область"
result = df[(df['indicator_code'] == indicator_code) & (df['object_name'] == region)]
print(result[['year', 'indicator_value', 'indicator_unit']].sort_values('year'))
```

```python
keywords_econ = ['отгру', 'оборот организаций', 'отгружен', 'произведен', 'произведено']
for kw in keywords_econ:
    found = df[df['indicator_name'].str.contains(kw, case=False, na=False)]
    n = found['indicator_code'].nunique()
    print(f"=== '{kw}': {n} индикаторов ===")
    if n > 0:
        print(found[['indicator_code', 'indicator_name']].drop_duplicates().head(5).to_string())
        print()
```

```python
keywords_shx = ['продукц', 'растениев', 'животнов', 'зерн', 'урожай', 'сельскохозяйствен']
for kw in keywords_shx:
    found = df[df['indicator_name'].str.contains(kw, case=False, na=False)]
    n = found['indicator_code'].nunique()
    print(f"=== '{kw}': {n} индикаторов ===")
    if n > 0:
        print(found[['indicator_code', 'indicator_name']].drop_duplicates().head(5).to_string())
        print()
```

```python
sections = df.groupby('section')['indicator_code'].nunique().sort_values(ascending=False)
print(f"Всего разделов: {len(sections)}")
print("\nТоп-30 разделов по числу индикаторов:")
print(sections.head(30).to_string())
```

```python
indicators = {
    'Отгрузка': 'Y477090007',
    'С/х продукция': 'Y477120029',
}

regions = ['Московская область', 'Краснодарский край', 'Республика Саха (Якутия)']

for name, code in indicators.items():
    print(f"\n=== {name} ({code}) ===")
    for region in regions:
        rows = df[(df['indicator_code'] == code) & (df['object_name'] == region)]
        if len(rows) == 0:
            print(f"  {region}: НЕТ ДАННЫХ")
        else:
            years = sorted(rows['year'].unique())
            last_year = max(years)
            last_value = rows[rows['year'] == last_year]['indicator_value'].iloc[0]
            unit = rows[rows['year'] == last_year]['indicator_unit'].iloc[0]
            print(f"  {region}: {min(years)}-{last_year}, последнее значение: {last_value} ({unit})")
```
